```python
#import libraries
# python packages  for audio video analysis
import librosa
import torch
import IPython.display as display
from transformers import Wav2Vec2ForCTC,Wav2Vec2Tokenizer
import numpy as np

```

```python
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
#audio,samppling_rate = librosa.load('/content/आईवर सादर केलेली कविता _ संकर्षण कऱ्हाडे _ Mother Poem _2 9M views · 47K reactions.m4a',sr= 16000)
audio,sampling_rate = librosa.load('/content/“Education is the... __ Nelson Mandela quotes.m4a',sr=16000)
```

```python
input_values = tokenizer(audio,return_tensors='pt').input_values
input_values
```

```python
logits = model(input_values).logits
logits
```

```python
predicted_ids = torch.argmax(logits,dim=-1)
```

```python
transcription= tokenizer.decode(predicted_ids[0])
```
